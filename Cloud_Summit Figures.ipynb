```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn-v0_8-colorblind')

SMALL_SIZE = 14
MEDIUM_SIZE = 20
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

marker = {'Batch Size 10': '',  'Batch Size 50': '', 'Batch Size 100': '','Batch Size 500': ''}
# Read data from CSV file
csv_file_path = '/content/drive/MyDrive/1.Improved Algorithm/Queues_Based_Batch Size Comparison/BatchSize_Throughput.xlsx'
data = pd.read_excel(csv_file_path)

# Assuming your CSV file has columns like 'Minutes', 'Batch Size 10', 'Batch Size 50', etc.
# Adjust column names accordingly
batch_sizes = ['Batch Size 10', 'Batch Size 50', 'Batch Size 100', 'Batch Size 500']

# Function to compute CDF
def compute_cdf(data):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    return sorted_data, cdf

# Convert milliseconds to minutes and scale to kilo-entries per minute
data_minutes = data.copy()
for col in batch_sizes:
    data_minutes[col] /= 1000  # Convert to seconds


# Plot CDF of throughput for each batch size
plt.figure(figsize=(6, 4))
for batch_size in batch_sizes:
    sorted_throughput, cdf = compute_cdf(data_minutes[batch_size])
    plt.plot(sorted_throughput, cdf, label=batch_size)

plt.legend(loc='upper right')
plt.xlabel('Throughput (k/min)')
plt.ylabel('CDF')

#plt.title('CDF of Throughput for Different Batch Sizes')
plt.legend(bbox_to_anchor=(1.37, 1.2), ncol=4, fontsize="12",)
#plt.grid(True)

plt.savefig('Throughput'+'.png', bbox_inches='tight')
#files.download(name+'.png')

plt.show()



```
